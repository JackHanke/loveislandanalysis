## Analysis

```python
import pandas as pd
import duckdb

LoveIslandResults = pd.read_csv('./data/LoveIslandResults.csv')
```

```python
print(f'How many contestants obtained each possible final status?')
duckdb.query('''
select status, count(*) freq from LoveIslandResults group by status order by freq
''').df()
```

```python
print(f'How many winners entered on each day?')
duckdb.query('''
select "Day Entered", count(*) freq from LoveIslandResults where status = 'Winner' group by "Day Entered" order by freq desc, "Day Entered"
''').df()
```

```python
print(f'What is the average day a contestant enters the villa?')
duckdb.query('''
select avg("Day Entered") from LoveIslandResults where status = 'Winner'
''').df()
```

```python
print(f'Who spent the longest time in the villa?')
duckdb.query('''
select * from LoveIslandResults where "Day Left"-"Day Entered" = (select max("Day Left" - "Day Entered") from LoveIslandResults)
''').df()
```

```python
print(f'How many days did each season run?')
duckdb.query('''
select country, season, max("Day Left") as NumDays from LoveIslandResults group by country, season order by country, season
''').df()
```
